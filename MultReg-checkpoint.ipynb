# Method Justification

### Summary of Assumptions

The outcome of multiple linear regression is a linear equation of the form

Y = b + a1x1 + a2x2 + … + anxn,

where Y is the target, or response variable, b is an intercept, the xi’s are the explanatory variables, and the ai’s are coefficients.  The equation is essentially a multi-variable analog of the common equation for a line in 2-dimensions: 

y = mx + b.  

Each of these elements should be numeric.  More specifically the predictive variable needs to be numeric and continuous, but the explanatory variables can be discrete and numeric.  The equation can be found by minimizing the differences between actual data points and a theorized best prediction or expectation.  Such expected values can be expressed in the same form as the linear equation above, where E(Y) is used to denote the expected value of the response variable.  Residuals are the differences between actual data points and the expected values.  The regression model depends on minimizing the set of residuals.  



### Tool Benefits

In a single variable model finding the smallest residuals can be found by hand using calculus, but the calculations become unwieldy with a larger number of variables.  Programming languages come in handy here.  Python has several pre-coded packages that contain useful functions for model creation and exploration.  Data sets can be stored in objects like data frames and arrays and used for input and output with these functions. # Model Comparison and Analysis

The statsmodels.formula.api package was used to create the initial multiple linear regression model using the ols() function.  This function specified the target and explanatory variables, and the fit() method was applied to it.  

```python
#should be v 21.x or above
import sklearn
print(sklearn.__version__)
```

```python
# define arrays for target and explanatory variables
y = mlr_df[['Tenure']]
X1 = mlr_df[['MonthlyCharge', 'Bandwidth_GB_Year', 'Outage_sec_perweek', 'Email', 'Yearly_equip_failure', 'Port_modem',  'Tablet', 'Phone', 'Multiple', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies','PaperlessBilling', 'Contract','InternetService', 'PaymentMethod']]
```

```python
# linear model on non-normalized data
from statsmodels.formula.api import ols
lin_mdl_v1 = ols('Tenure ~ MonthlyCharge + Bandwidth_GB_Year + Outage_sec_perweek + Email + Yearly_equip_failure + Port_modem + Tablet + Phone + Multiple + OnlineSecurity + OnlineBackup + DeviceProtection + TechSupport + StreamingTV + StreamingMovies + PaperlessBilling + Contract + InternetService + PaymentMethod', data = mlr_df).fit()
```

### Model Reduction

Applying the summary() method to the model produces coefficients for each explanatory variable and their corresponding P-values.  The P-values quantifies how much of an effect changing a predictor variable will have on the response variable under a null hypothesis that a coefficient is 0, which means there is no effect.  The commonly used low value of < 0.05 is used as a cutoff.  This means that if P > 0.5 it’s not statistically significant to rule out a 0 coefficient, and thus such a variable should be removed from the model.  

```python
# look at P values to weed out variables with higher uncertainty in the model: cut-off P>0.05
lin_mdl_v1.summary()
```

```python
# updated dataframe 
mlr_df2 = mlr_df.drop(['Outage_sec_perweek', 'Email', 'Yearly_equip_failure', 'Port_modem', 'Tablet', 'Phone', 'PaperlessBilling', 'Contract', 'PaymentMethod'], axis = 1)
mlr_df2.head()
```

An improved model was created upon removing the columns with corresponding P-values > 0.05.  

```python
# refined model
lin_mdl_v2 = ols("Tenure ~ MonthlyCharge + Bandwidth_GB_Year + Multiple + OnlineSecurity + OnlineBackup + DeviceProtection + TechSupport + StreamingTV + StreamingMovies + InternetService",
                data = mlr_df2).fit()
```

```python
lin_mdl_v2.summary()
```

Variance inflation factors are used as a measure of collinearity.  This is where one or more explanatory variables have a correlation with each other.  (Zach, 2020) This can be problematic if one variable predicts another when the goal is to get multiple variables to predict just one.   The code that produces the VIF metrics uses the statsmodels library and was adapted from the statology.org site.  Dr. Sewell recommends rejecting any variable with a VIF > 10 (slide 22), but the Statology article suggests correlation issues when VIF > 5.  

```python
# look at variance inflation factors - VIF's - to check for collinearity
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor

y, X = dmatrices("Tenure ~ MonthlyCharge + Bandwidth_GB_Year + Multiple + OnlineSecurity + OnlineBackup + DeviceProtection + TechSupport + StreamingTV + StreamingMovies + InternetService", data=mlr_df2, return_type='dataframe')

# calculate VIF for each explanaroty variable
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['variable'] = X.columns

vif
```

MonthlyCharge has a VIF of 7.6, which may pose issues with collinearity.  It was dropped from the updated dataframe and a 3rd model was created.

```python
# update dataframe again
mlr_df3 = mlr_df.drop(['MonthlyCharge'], axis = 1)
mlr_df3.head()
```

```python
# refine model again
lin_mdl_v3 = ols("Tenure ~ Bandwidth_GB_Year + Multiple + OnlineSecurity + OnlineBackup + DeviceProtection + TechSupport + StreamingTV + StreamingMovies + InternetService",
                data = mlr_df3).fit()
lin_mdl_v3.summary()
```

```python
# Checking VIF again

y, X = dmatrices("Tenure ~ Bandwidth_GB_Year + Multiple + OnlineSecurity + OnlineBackup + DeviceProtection + TechSupport + StreamingTV + StreamingMovies + InternetService", data=mlr_df2, return_type='dataframe')

# calculate VIF for each explanaroty variable
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['variable'] = X.columns

vif
```

lin_mdl_v3 resulted in no columns with P-values > 0.05, and all VIFs are small enough to ease suspicion of collinearity.  This is the final model.

# Final Model: Calculations and Comparisons

The strength of the model can be measured by correlation coefficient R^2, which was 0.995 in the final model.  This is very good, as the closer R^2 is to 1 the stronger the fit of the model.  The intercept and coefficients for the model can be obtained from the output of its summary.  If  y represents the target variable and the xi’s represent the column names in order of appearance in the summary output, the equation of fit can be expressed as:

Y = -0.7559 + 0.0122x1 - 0.8183x2 - 0.9233x3 - 1.1208x4 - 1.0274x5 - 0.0809x6 – 2.7852x7 – 2.5482x8 - 2.7932x9.


The initial model had an ever so slightly higher R^2 of 0.997, but contained 9 more explanatory variables whose coefficients likely had negligible effect on the target variable according to their high P-values.   VIFs were computed on the original model as well, and only MonthlyCharge seemed like it may possibly have a correlation with another explanatory variable.  After removing it the VIFs went down to slightly above 1 for all input variables.  Removing this column also did not have a strong enough effect on P-values to warrant further eliminations. Reducing the model from 19 to 9 predictor variables seems to have gone a long way to mitigate overfitting.  If the model were dependent on the higher number of features, then changing a portion of them could have a more profound effect on the target than is desirable.  With less explanatory variables the response variable is more likely to change in a way that doesn’t vary too widely from its observed range.

The last point above ties into residuals, which give a measure of difference between observed and expected values.  The value for observed response variable which corresponds with a specific observation of a single explanatory variable is compared with the output from the model that is produced from that observed explanatory variable.  The numerical differences are the residuals and can be plotted along a line that the target variable follows for all of the other explanatory variables.  This line is called a zero line, and a good model will have a symmetric distribution of residuals about this line.  The “Residuals versus (input variable)” is the graph that shows this characteristic.  The plots for each explanatory variable display a very equal looking proportion of residual points above and below the zero-line.  This is another indication of the final model’s accuracy.

# Results

Referring to the equation of fit for the second model, it is apparent that the coefficient for MonthlyCharge is positive, which indicates that if a customer’s bill was increased, they may stay longer.  This seems counter-intuitive, since people may be more inclined to switch providers if they have less expensive offerings.  On the other hand, inflation causes prices to increase, so increased charges over time is to be expected.  This variable may have caused collinearity issues, and by dropping it the above concerns are no longer present.  Also, the range of Bandwidth_GB_Year is much wider than the other variables, and the relatively small magnitude of its coefficient seems to compensate for that when compared to the others.  


The coefficients for the rest of the categorical explanatory variables are much larger, but negative.  The range of their values is much smaller, so their coefficients seem to compensate for that by having a larger magnitude.  TechSupport offering seems to have the smallest of this effect on Tenure, but overall, the more customers who have these additional services the lower the average customer stay will be according to the model.  The two Streaming variables have the largest negative effect.  This may point to usage of independent streaming services like Netflix, or perhaps that such services offer a better experience than what is offered from the ISP.  InternetService also has a relatively large negative coefficient, indicating that the more customers who have it the shorter the average Tenure is.

### Recommendations

The relatively significant effect of InternetService on Tenure seems a little baffling.  It would seem to make sense that if more customers have internet service as part of their package, they may keep their services longer overall.  Perhaps there is a large overlap in service area where competitor service is available and internet pricing is a point of fickleness with consumer behavior.  It may also be that this ISP has traditionally offered cable or phone services, while internet is newer and not up to par in terms of quality.  This regression model cannot reveal such insight, but only reveals the trend.  Perhaps if this variable were used in a different predictive model more insight may be possible.  In general, the other optional services: OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport, StreamingMovies, and Streaming video all seem to have a negative effect on Tenure.  According to the model customers would keep their services longer overall if less had such add-ons.  Perhaps focusing less on promoting such offerings or even doing away with some of them all together could benefit the bottom line by allowing the ISP to hold on to customers longer.
